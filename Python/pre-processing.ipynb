```python
def GetComments(id,cursor=1683600120000):
    url_ = "http://127.0.0.1:3000/comment/new?id=" + str(id) + "&sortType=3&cursor=" + str(cursor) + "&type=0&pageSize=500&pageNo=2"
    data_ = requests.request("GET", url_).json()
    COLUMNS = ["user.userId","commentId", "user.nickname", "content", "timeStr", "likedCount", 'ipLocation.location', 'time']
    df_ = json_normalize(data_['data']["comments"])
    return df_[COLUMNS]
```

```python
GetComments(2026224214)
```

```python
def get_all_comments(id,iteration,cursor=1683600120000):
    ans = GetComments(id, cursor)
    time = re.sub(u"([^\u0030-\u0039])", "", str(ans.iloc[-1].time))
    for i in range(iteration):
        new_df = GetComments(id, time)
        time = re.sub(u"([^\u0030-\u0039])", "", str(new_df.iloc[-1].time))
        ans = ans.append(new_df)
    return ans
```
